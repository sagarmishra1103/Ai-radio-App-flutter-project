```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/Metadata_Country.csv")
data
```

```python
data.describe()
```

```python
df1=data[["Country Code","Region","IncomeGroup","TableName"]]
df1
```

```python
df1.dropna(subset=['Region'], inplace=True)
df1.rename(columns={'TableName': 'CountryName'}, inplace=True)
df1
```

```python
#bar chart plotting
Region_counts = df1['Region'].value_counts()
print(f'\ncounts of countries by Region :\n\n{Region_counts}\n')
x=range(len(Region_counts.index))

plt.bar(Region_counts.index,Region_counts.values,color='#6FC276')
plt.xlabel('Region groups')
plt.ylabel('Region Counts')
plt.title('Categorical Distribution of Region')


plt.xticks(x,Region_counts.index,rotation=45,color='#BBB5D8')
plt.legend(title='Region')
plt.tight_layout()
plt.show()
```

```python
#bar chart plotting
IncomeGroup_counts = df1['IncomeGroup'].value_counts()
print(f"\nCounts of Countries by IncomeGroup:\n\n{IncomeGroup_counts}\n")

#plot the bar chart
x=range(len(IncomeGroup_counts.index))
plt.bar(IncomeGroup_counts.index,IncomeGroup_counts.values,color='#FF7F7F')
plt.xlabel('Income Groups')
plt.ylabel('Count')
plt.title('Categorical Distribution of IncomeGroup')

plt.xticks(x,IncomeGroup_counts.index,rotation=45,color='#BBB5D8')
plt.legend(title='IncomeGroup')
plt.tight_layout()
plt.show()
```

```python
# Syntax for creating a 2D categorical distribution
categorical_distribution = pd.crosstab(df1['Region'], df1['IncomeGroup'])

# Plotting the bar plot using pandas' plot() function
categorical_distribution.plot(kind='bar', stacked=True, figsize=(8, 6))

# Customize the plot (optional)
plt.title('2-D Categorical Distribution: Region vs. IncomeGroup')
plt.xlabel('Region')
plt.ylabel('Count')
plt.legend(title='IncomeGroup')

# Show the plot
plt.tight_layout()
plt.show()
```
